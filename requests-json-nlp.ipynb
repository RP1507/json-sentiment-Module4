The following code accesses the [lyrics.ovh](https://lyricsovh.docs.apiary.io/#reference/0/lyrics-of-a-song/search) public api, searches for the lyrics of a song, and stores it in a dictionary object.
```python
import requests
import json
import xmltodict  #had to install this for sites xml results / was getting error when trying to use api

#result = json.loads(requests.get('https://api.lyrics.ovh/v1/They Might Be Giants/Birdhouse in your soul').text) ----- This api is not working, going to try the chartlyrics API

result = requests.get('http://api.chartlyrics.com/apiv1.asmx/SearchLyricDirect?artist=They Might Be Giants&song=Birdhouse in your soul').text
      
# convert sites results (xml) to JSON dict
d = xmltodict.parse(result)
xmlstr = json.dumps(d)  
data = json.loads(xmlstr)

#Write results to file 
with open("BirdhouseInYourSoul.json", "w") as song:
    json.dump(data, song)
    
print(result)
```

##### 2. Read in the contents of your file.  Print the lyrics of the song (not the entire dictionary!) and use spaCyTextBlob to perform sentiment analysis on the lyrics.  Print the polarity score of the sentiment analysis.  Given that the range of the polarity score is `[-1.0,1.0]` which corresponds to how positive or negative the text in question is, do you think the lyrics have a more positive or negative connotaion?  Answer this question in a comment in your code cell.

### Question 2:

```python
import json
import spacy
from spacytextblob.spacytextblob import SpacyTextBlob

# Open the saved JSON file
with open('BirdhouseInYourSoul.json') as song:
    # Reading from json file
    song_data = json.load(song)
    
print("Lyrics: ", song_data['GetLyricResult']['Lyric'])

# get polarity with spaCy/text blobs
nlp = spacy.load('en_core_web_sm')
lyrics = str(song_data['GetLyricResult']['Lyric'])
nlp.add_pipe("spacytextblob")
doc = nlp(lyrics)

# Get and print the polarity score of the song
   
print("\n\nThe polarity score for",song_data['GetLyricResult']['LyricSong'],"is:", doc._.blob.polarity)    

##This songs polarity score is fairly neutral. It is slightly above 0 (towards the positive side) but falls in the range of -0.05 to 0.05. So it's polarity scores is neutral. 
```
